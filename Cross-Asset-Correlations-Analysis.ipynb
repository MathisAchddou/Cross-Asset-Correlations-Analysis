```python
import numpy as np
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
tickers = {'ES=F': 'S&P 500', 'NQ=F': 'Nasdaq 100', 'RTY=F': 'Russel 2000', 'YM=F': 'Dow Jones',
           'CL=F': 'Crude Oil', 'NG=F': 'Natural Gas',
           'GC=F': 'Gold', 'HG=F': 'Copper', 'SI=F': 'Silver',
           'DX=F':'US Dollar', '6E=F': 'Euro', '6J=F': 'Japanese Yen', '6B=F': 'British Pound', '6A=F': 'Australian Dollar', '6C=F': 'Canadian Dollar', '6N=F': 'New Zealand Dollar', '6S=F': 'Swiss Franc',
           'BTC=F': 'Bitcoin', 'ETH=F': 'Ethereum',
           'ZN=F': '10-Year T-Note', 'ZF=F': '5-Year T-Note', 'ZT=F': '2-Year T-Note', 'ZB=F': 'U.S. Treasury Bond',
           'ZC=F': 'Corn', 'ZS=F': 'Soybean', 'ZW=F': 'Wheat'}

df = pd.DataFrame()
```

```python
for key, value in tickers.items():
    data = yf.download(key, period = 'max')
    df[value] = data['Close'].pct_change() * 100    
```

```python
df = df.reset_index().dropna()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

```python
plt.figure(figsize=(22,22))
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
heatmap = sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Cross Asset Correlation heatmap since Feb 2021',pad=20);
```

```python
def filter_data(df, period):
    if period != 0:
        current_date = pd.Timestamp.now()
        offset = current_date - pd.DateOffset(months=period)
        df = df[df['Date'] > offset]

    df = df.drop(['Date'], axis=1)
    df = df.reset_index(drop=True)
        
    return df
            
df_1m = filter_data(df, 1)
df_3m = filter_data(df, 3)
df_6m = filter_data(df, 6)
df_12m = filter_data(df, 12)
df_all = filter_data(df, 0)

dfs = {'One Month':df_1m, 'Three Months':df_3m, 'Six Months':df_6m, 'Twelve Months':df_12m, 'Feb 2021':df_all}


all_df = [df_1m, df_3m, df_6m, df_12m, df_all]
```
